## Prints a spiral 3d image using parametric equations

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
# %matplotlib # for nice presentation and move 3d object with mouse
```

```python
x=[];y=[];z=[]
# create array of angles each with 3.14/360 degrees
# array stores 2 circles (4*3.14)
angle_range = np.arange(0,4*np.pi+(np.pi/360),(2*np.pi)/360)
z_step=0
for angle in angle_range:
    # parametric equations x = cos(degrees (in radians)),
    #                      y = sin(degrees (in radians))
    x.append(np.cos(angle))
    y.append(np.sin(angle))
    z.append(z_step)
    z_step+= 2
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
mpl.rcParams['legend.fontsize'] = 20
ax.plot(x, y, z, label='circle enters the 3rd dimension!')
font = {'family': 'Suruma',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }
ax.set_xlabel("x",fontdict=font)
ax.set_ylabel("y",fontdict=font)
ax.set_zlabel("z",fontdict=font)
ax.legend()
plt.show()
```
